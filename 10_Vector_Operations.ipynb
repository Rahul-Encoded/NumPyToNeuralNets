```python
# Dot Product of 2 vectors
a = np.array([2,3])
b = np.array([4,4])

a_b = np.dot(a,b) #a.b = |a||b|costheta
print(a_b)
```

```python
# program to generate random numbers
import random
def random_generator(count, min_value, max_value):
	rand_arr = []
	while(count > 0):
	  rand_x = random.randint(min_value, max_value)
	  rand_arr.append(rand_x)  # Just append, don't reassign
	  count -= 1
	return rand_arr

random_generator(3, 0, 100)
```

```python
c = np.array(random_generator(3, 0, 100))
d = np.array(random_generator(3, 0, 100))
print(c)
print(d)

c_d = np.dot(c,d)
print(c_d)
```

```python
# Cross product of 2 vectors
axb = np.cross(a,b)
print(axb)
```

```python
cxd = np.cross(c,d)
print(cxd)
```
