## Label Encoding
- converting the labels into numeric form

```python
# importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding of Breast Cancer Dataset
# loading the data from csv to pandas dataframe
cancer = pd.read_csv("Datasets\\breast_cancer.csv")
```

```python
# load the label Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer.diagnosis)
```

```python
# appending the labels to the DataFrame
cancer["target"] = labels
```
