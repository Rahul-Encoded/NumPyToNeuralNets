### Useful for making Plots

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
import math
```

```python
x = np.linspace(0, 2*math.pi, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x, y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#ading title, x-axis and y-axis labels
plt.plot(x, y)
plt.xlabel("angle in radians")
plt.ylabel("sine value")
plt.title("sine wave")
plt.show()
```

```python
#parabola
x = np.linspace(-10, 10, 20)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, "r+")
plt.show()
```

```python
plt.plot(x,y,"g.")
plt.show()
```

```python
plt.plot(x,y,"bx")
plt.show()
```

```python
x = np.linspace(-10, 10, 150)
plt.plot(x , np.sin(x), "g-")
plt.plot(x , np.cos(x), "r--")
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])	#[c, y, width, height,]
lang = ["English", "French", "Spanish", "Latin", "Hindi"]
people = [100, 20, 30, 40, 150]
ax.bar(lang, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NUMBER OF PEOPLE")
plt.show()
```

Pie-Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])	#[c, y, width, height,]
lang = ["English", "French", "Spanish", "Latin", "Hindi"]
people = [100, 20, 30, 40, 150]
ax.pie(people, labels = lang, autopct="%1.1f%%")
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color="g")
ax.scatter(x, z, color="r")
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection="3d")
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap="Greens")
plt.show()
```
