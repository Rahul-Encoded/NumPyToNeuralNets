## Data Visualization library

```python
#Importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#setting a theme for plots
sns.set_theme()
# visualize the dataset
sns.relplot(data = tips, x = "total_bill", y="tip", col="time", hue = "smoker", style="smoker", size="size")
```

House price dataset

```python
# Importing the California housing dataset
from sklearn.datasets import fetch_california_housing

# Load the California housing dataset
housing = fetch_california_housing()
```

```python
house = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
house["PRICE"] = housing.target
```

```python
house.head()
```

Distribution plot

```python
sns.displot(house["PRICE"])
```

```python
sns.displot(house["PRICE"], kde=True)
```

```python
sns.histplot(house["PRICE"], kde=True)
```

Correlation:
- Positive
- Negative

Heat Map

```python
# Constructing a Heat Map
correlation = house.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt=".1f", annot=True, annot_kws={"size":8})
```
